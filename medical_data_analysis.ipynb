```python
#PLAYING WITH LISTS OF DATA 
#using our new knowledge of Python strings to obtain and clean up medical data so that it is easier to read and analyze.

#The string medical_data stores the medical records for ten individuals. 
#Each record is separated by a ; and contains the name, age, BMI (body mass index), and insurance cost for an individual, in that order
medical_data = \
"""Marina Allison   ,27   ,   31.1 , 
#7010.0   ;Markus Valdez   ,   30, 
22.4,   #4050.0 ;Connie Ballard ,43 
,   25.3 , #12060.0 ;Darnell Weber   
,   35   , 20.6   , #7500.0;
Sylvie Charles   ,22, 22.1 
,#3022.0   ;   Vinay Padilla,24,   
26.9 ,#4620.0 ;Meredith Santiago, 51   , 
29.3 ,#16330.0;   Andre Mccarty, 
19,22.7 , #2900.0 ; 
Lorena Hodson ,65, 33.1 , #19370.0; 
Isaac Vu ,34, 24.8,   #7045.0"""


def clean_medical_data( medical_data):
  # We want the insurance costs to be represented in US dollars.
  #Replacing all instances of # in medical_data with $
  updated_medical_data = medical_data.replace("#","$")
  #print(updated_medical_data)

#to calculate the number of medical records in our data
  num_records = 0
# a for loop to iterate through the updated_medical_data string
  for character in updated_medical_data:
      #adding 1 to num_records when the current character is equal to $
    if character == '$':
      num_records+=1

  #print("There are "+ str(num_records)+" medical records in the data.")
    
#Splitting Strings
    
    #plitting the updated_medical_data string 
    #into a list of each medical record. Remember that each medical record is separated by a ; in the string.
  medical_data_split  = updated_medical_data.split(";")
  

  medical_records = []

  for record in medical_data_split :
    for character in record: 
      if character == ",":
         medical_records.append(record.split(","))

  #print(medical_records)
  medical_records_clean = []
# split each medical record into its own list because so far they are hard to read
  for rec in medical_records:
      #creating an empty list called record_clean. Weâ€™ll use this list to store a formatted version of each medical record.
    record_clean = []
      #for loop that goes through each record:
    for item in rec:
      record_clean.append(item.strip())
        #add each cleaned up record to medical_records_clean
    medical_records_clean.append(record_clean)
    medical_records_clean_new = []
#removing the duplicate records that were previously created by accident
    for x in medical_records_clean:
      if x not in medical_records_clean_new:
        medical_records_clean_new.append(x)

          #Analyzing Data

#data is now clean and ready for analysis

    #printing out the names of each of the ten individuals
  for record in medical_records_clean_new:
      #want all of the names in the medical records to be in uppercase characters
    record[0]=record[0].upper()
    print(record[0])

    
    

  #new separate lists to store the data separately
  names = []
  ages = []
  bmis = []
  insurance_costs = []


  #dividing the data into different lists
  for record in medical_records_clean_new:
    # append the name, age, BMI, and insurance cost from the current medical record 
    names.append(record[0])
    ages.append(record[1])
    bmis.append(record[2])
    insurance_costs.append(record[3])


  print(names)
  print(ages)
  print(bmis)
  print(insurance_costs)


#1st function call
clean_medical_data( medical_data)

bmis = ['31.1', '22.4', '25.3', '20.6', '22.1', '26.9', '29.3', '22.7', '33.1', '24.8']
insurance_costs = ['31.1', '22.4', '25.3', '20.6', '22.1', '26.9', '29.3', '22.7', '33.1', '24.8']


#Now that all of our data is in separate lists, we can easily perform analysis on that data. ->find the average

def calculate_average_from_lists(bmis, insurance_costs):
  total_bmi = 0
  total_insurance_cost=0
  for bmi in bmis:
    total_bmi+=float(bmi)
  average_bmi = 0
  for insurance_cost in insurance_costs:
    insurance_cost.replace("$"," ")
    insurance_cost.strip()

  insurance_costs_float = [float(ic) for ic in insurance_costs ]
  for ic in insurance_costs_float:
    total_insurance_cost +=ic
  average_insurance_cost=0
  
  average_insurance_cost=total_insurance_cost/len(insurance_costs_float)
  

  average_bmi  = total_bmi /len(bmi)
  print("Average bmi : " + str(average_bmi ))
  print("Average insurance cost : " + str(average_insurance_cost))

  #2nd function call
calculate_average_from_lists(bmis, insurance_costs)
```
